**5 parts of the data**

1.   Game scores: Information such as the winning team, game duration, rest days since the previous game, and whether it was a home or away game for this season's games (2024-25)
2.   Player scores per game: Individual player performance metrics for this season (2024-25), including points, field goal percentage, steals, blocks, etc.
3.   Player metadata: Information on players' height, weight, and years in the league
4.   Player last season data: Information on minutes played performance in the previous season, and if the player was traded in the last season (2023-24)
5.   Player this season data: Information on players' age and if the player was traded in this season (2024-25)

Box Score


*   Total 9635 player's boxsores in season 2024-25 as of 1/5/2025
*   Keep only representative variables
  * FG_PCT: Field Goal Percentage = Field Goals Made / Field Goals Attempted (including both 2 pointers and 3 pointers)
  * FT_PCT: Free Throw Percentage = Free Throws Made / Free Throws Attempted
  * REB: Total Rebounds including both OREB (Offensive Rebound) and DREB (Defensive Rebound)
  * AST: Assit occurs when a player who passes the ball to a teammate and that lead to a field goal
  * STL: Steal occurs when a defensive player takes the ball way from an offensive player
  * BLK: Block occurs when a defensive player deflects a field goal attempt
  * TO: Turnover occurs when losing the ball posession before a shot is attempted due to bad passes, traveling, any fouls - the more TO the worse
  * PF: Personal fouls occurs when players made an illegal contact with an opponent player
  
*   REB seems highly correlated to other stats
